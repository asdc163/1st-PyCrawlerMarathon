### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.ettoday.net/news/news-list.htm')
response = r.text
soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
soup.find_all("div", class_="part_list_2")
```

```python
dic = {}
dic['影劇'] = soup.find_all('em', class_="tag c_star")
dic['生活'] = soup.find_all('em', class_="tag c_lifestyle")
dic['時尚'] = soup.find_all('em', class_="tag c_fashion")
dic['政治'] = soup.find_all('em', class_="tag c_news")
dic['房產雲'] = soup.find_all('em', class_="tag c_house")
dic['財經'] = soup.find_all('em', class_="tag c_finance")
dic['旅遊'] = soup.find_all('em', class_="tag c_travel")
dic['社會'] = soup.find_all('em', class_="tag c_society")
dic['論壇'] = soup.find_all('em', class_="tag c_commentary")
dic['大陸'] = soup.find_all('em', class_="tag c_china")
dic['體育'] = soup.find_all('em', class_="tag c_sport")
dic['寵物動物'] = soup.find_all('em', class_="tag c_pet")
dic['保險'] = soup.find_all('em', class_="tag c_insurance")
dic['遊戲'] = soup.find_all('em', class_="tag c_game")
dic['健康'] = soup.find_all('em', class_="tag c_health")
dic['ET車雲'] = soup.find_all('em', class_="tag c_speed")
dic['3C家電'] = soup.find_all('em', class_="tag c_teck3c")
dic['國際'] = soup.find_all('em', class_="tag c_global")
dic['影劇'] = soup.find_all('em', class_="tag c_star")
dic['地方'] = soup.find_all('em', class_="tag c_local")
```
